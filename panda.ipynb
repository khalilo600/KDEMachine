```python
# pandas
import pandas as pd

# Create a DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}
df = pd.DataFrame(data)

# Display the DataFrame
print("DataFrame:")
print(df)

# Accessing columns
print("\nAccessing a column:")
print(df['Name'])

# Accessing rows
print("\nAccessing a row:")
print(df.iloc[1])

# Adding a new column
df['Salary'] = [70000, 80000, 90000]
print("\nDataFrame after adding a new column:")
print(df)

# Removing a column
df.drop('City', axis=1, inplace=True)
print("\nDataFrame after removing a column:")
print(df)

# Summary statistics
print("\nSummary statistics:")
print(df.describe())

```
